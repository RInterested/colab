```
options(repr.plot.width=5, repr.plot.height=5)

x <- c(570, 585, 591, 586, 631, 611, 617, 622, 622)
y <- c(693, 683, 673, 663, 653, 633, 623, 613, 603)
plot(x,y, pch=19)
mod <- lm(y ~ x)
abline(mod, col = 'firebrick')
summary(mod)
points(x[5],y[5],col="red",pch=19,cex=2)
legend(605,665,legend="Leverage + outlier", bg="transparent", text.col=2, box.lty=0)
```

```
options(repr.plot.width=6, repr.plot.height=6)
par(mfrow=c(2,2))
plot(mod, pch=19, cex=.7)
```

```
new.x <- x[-5]
new.y <- y[-5]
plot(new.x,new.y, pch=19)
mod2 <- lm(new.y ~ new.x)
abline(mod2, col = 'firebrick')
summary(mod2)
par(mfrow=c(2,2))
plot(mod2)
```
