The way I understand conditional expectation and teach my students is the following:

***conditional expectation $E[Y|\sigma(X)]$ is a picture taken by a camera with resolution $\sigma(X)$***

As mentioned by Alecos Papadopoulos, the notation $E[Y|\sigma(X)]$ is more precise than $E[Y|X]$. Along the line of camera, one can think of $Y$ as the original object, e.g., a landscape, scenery. $E[Y|\sigma(X,Z)]$ is a picture taken by a camera with resolution $\sigma(X,Z)$. Expectation is an averaging operator ("blurring" operator?). The scenary may contain a lot of stuff, but the picture you took using a camera with low resolution will certainly make some detail go away, e.g., there may be an UFO in the sky that can be seen by your naked eye but it does not appear in your picture taken by (iphone 3?)

If the resolution is so high such that $\sigma(X,Z)=\sigma(Y)$, then this picture is able to capture every detail of the real scenery. In this case, we have $E[Y|\sigma(Y)]=Y$.

Now, $E[E[Y|\sigma(X,Z)]|\sigma(X)]$ can be viewed as: ***using another camera with resolution $\sigma(X)$ (e.g., iphone 1) which is lower than $\sigma(X,Z)$ (e.g., iphone 3) and take a picture on that picture generated by camera with resolution $\sigma(X,Z)$***, then it should be clear that ***this picture on a picture should be the same as if you originally just use a camera with low resolution $\sigma(X)$ on the scenery.***

This provides intuition on $E[E[Y|X,Z]|X]=E[Y|X]$. ***In fact this same intuition tells us that $E[E[Y|X]|X,Z]=E[Y|X]$ still. This is because: if your first picture is taken by iphone 1 (i.e., low resolution), and now you want to use a better camera (e.g., iphone 3) to generate another photo on the first photo, then there is no way you can improve the quality of the first photo.***



I want to bring home the point about the camera resolution in @KevinKim's post. The "picture" makes reference to the expectation or probability-weighted average of a random variable $Y$ - a measure of centrality, and indeed a snapshot or summary of sorts of it.

Conditioning the expectation of $Y$ to a different random variable $X$ makes the expectation a function of $X,$ and while conditioning on $Y$ itself will return the entire support of the random variable, the sigma algebra generated by $X$ will provide a more or less detailed summary of $Y$ when conditioning the expectation on $X.$ So far just paraphrasing Kevin's response, but I wanted to see it in a very concrete example:

Say that  $\small Y\sim \begin{pmatrix}I&II&III&IV&V&VI\\1&2&3&4&5&6\end{pmatrix}$ is the discrete random variable that simply records the value of a Roman numeral single unbiased die.

And I set out to prove that $$\small \mathbb E\left [ \mathbb E(Y \vert X,Z)\, \vert \, X \right]= \mathbb E[Y\,\vert \,X]\tag 1$$ as desired, and also, to see how the "picture" differs in resolution when instead I swap the order of the conditional part to  $$\small \mathbb E\left [ \mathbb E(Y \vert X)\, \vert \, X,Z \right]= \mathbb E[Y\,\vert \,X,Z]\tag 2$$

The random variable $\small X\sim \begin{pmatrix}I&II&III&IV&V&VI\\0&1&0&1&0&1\end{pmatrix}$ is just the dummy variable coding whether the outcome of the die toss is odd or even.

Similarly, the random variable $\small Z\sim \begin{pmatrix}I&II&III&IV&V&VI\\0&1&1&0&1&0\end{pmatrix}$ is the dummy variable coding for primality of the outcome.

The atoms in the generated sigma algebras should be as follows:

For $\sigma(Y)$ the atomic events are the sets $\small \left\{ \left \{I\right\} , \left \{II\right\} , \left \{III\right\} , \left \{IV\right\} , \left \{V\right\} , \left \{VI \right\}   \right\}.$

For $\sigma(X)$ the atomic events are the sets $\small \left\{ \left \{I,III,V \right\} , \left \{II,IV,VI \right\}  \right\}.$

For $\sigma(Z)$ the atomic events are the sets $\small \left\{ \left \{I,IV,VI \right\} , \left \{II,III,V \right\}  \right\}.$

For $\sigma(X,Z)$ the atomic events are the sets $\small \left\{ \left \{I\right\} , \left \{II\right\} , \left \{III,V\right\} , \left \{IV,VI\right\}   \right\},$ because $\small \{I \}$ will be the outcome that will be mapped to $x=0$ and $z=0,$ or $x,z=(0,0);$ whereas $\small \{II \}$ will be mapped to $x,z=(1,1);$ both $\small\{III\}$ and $\small\{V\},$ being prime and odd will be indistinguishably mapped to $x,z=(0,1);$ and $\small\{IV\}$ and $\small\{VI\}$ will be mapped to $x,z=(1,0),$ being composite and even.

The RHS of $(1)$ is the expectation of $Y$ given that the outcome is either odd or even.

If it is odd $x=0,$ the possible outcomes are naturally the in the first atomic subset above, $\small \{I,III,V\}$ with each one of these possible outcomes constituting the event 'odd' being mapped with a probability of one-third to the corresponding $y=1,$ $y=3$ and $y=5$ for an expectation $\mathbb E(Y\,\vert \, x=0)=\frac {1}{3} 9=3.$

If it is even $x=1,$ we will be dealing with any of the outcomes in the subset  $\small \{II,IV,VI\},$ which will be mapped by $Y$ to  $y=2,$ $y=4$ and $y=6$ with equal probability and for  $\mathbb E(Y\,\vert \, x=1)=\frac {1}{3} 12=4.$

So this camera snap shot has reduced the initial $Y$ to just two values:

But before moving on to taking a better picture, the LHS of $(1)$ or $\small \mathbb E\left [ \mathbb E(Y \vert X,Z)\, \vert \, X \right]$ would follow a similar logic: when $x=0,$ or the number is odd, the possible outcomes are $\small\{I,III,V\},$ but in the sigma algebra generated by both random variables $\sigma(X,Z)$ only two events are possible: either the outcome is odd and not prime, i.e. $\small\{I\},$ or the more common combination of odd and prime, i.e. $\small\{III,V\},$ which would occur respectively with $\frac 1 3$ and $\frac 2 3$ probabilities. In the first instance, the  $\mathbb E(Y \vert x,z=(0,0))=1;$ while in the second instance,  $\mathbb E(Y \vert x,z=(0,1))=\frac 1 2 (3+5)=4.$ Hence  $\small \mathbb E\left [ \mathbb E(Y \vert X,Z)\, \vert \, x=0 \right]=\frac 1 3 \cdot 1 + \frac 2 3 \cdot 4=3.$ A parallel reasoning for even outcomes paired with prime $\small\{II\}$ and composite $\small\{IV,VI\}$ events in the sigma algebra of $Z$ result in $\small \mathbb E\left [ \mathbb E(Y \vert X,Z)\, \vert \, x=1 \right]=\frac 1 3 \cdot 2 + \frac 2 3 \cdot 5=4.$

Now, for a higher-res picture, I turn to LHS of equation $(2)$ to find the possible atomic events corresponding to the combinations of odd/even with prime/composite to elicit the following conditional expectations:

$\small \mathbb E\left [ \mathbb E(Y \vert X)\, \vert \, x,z=(0,0) \right]=1,$ corresponding to the event $\small\{I\}$ with $y=1.$

$\small \mathbb E\left [ \mathbb E(Y \vert X)\, \vert \, x,z=(1,1) \right]=2,$ corresponding to the event $\small\{II\}$ with $y=2.$

$\small \mathbb E\left [ \mathbb E(Y \vert X)\, \vert \, x,z=(0,1) \right]=\frac 1 2 (3+5)=4,$ corresponding to the event $\small\{III,V\}$ with $y=3$ or $y=5.$

$\small \mathbb E\left [ \mathbb E(Y \vert X)\, \vert \, x,z=(1,0) \right]=\frac 1 2 (4+6)=5,$ corresponding to the event $\small\{IV,VI\}$ with $y=4$ or $y=6.$
