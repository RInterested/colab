[Google trends 'soccer' searches in the US](https://trends.google.com/trends/explore?date=all&geo=US&q=soccer)

```
install.packages('RCurl') # Needed to import csv from the GitHub repository.
require(RCurl)
```

```
x = getURL("https://raw.githubusercontent.com/RInterested/DATASETS/gh-pages/interest%20in%20soccer%20in%20the%20US.csv")
dat <- read.csv(textConnection(x))
dat$date <- as.Date(paste(dat$date,"-01",sep=""), format="%Y-%m-%d")
head(dat)[1:3,]
```

```
options(repr.plot.width=21, repr.plot.height=8)

plot(dat, type='l', xaxt='n', lwd=3, col='firebrick', xlab='', ylab='', main="Google searches in the US including 'soccer': 4-year seasonality")
at1 <- seq(min(dat$date), max(dat$date)+10, by="3 months")
axis.Date(1, at=at1, format="%b %d", las=2, cex.axis=0.7)
legend(dat$date[order(dat$count, decreasing=T)[1]] - 300, 80, legend=c("Germany World Cup"), bty='n')
legend(dat$date[order(dat$count, decreasing=T)[3]] - 300, 80, legend=c("S. Africa World Cup"), bty='n')
legend(dat$date[order(dat$count, decreasing=T)[2]] - 300, 80, legend=c("Brazil World Cup"), bty='n')
legend(dat$date[order(dat$count, decreasing=T)[7]] - 300, 80, legend=c("Russia World Cup"), bty='n')
abline(v = dat$date[order(dat$count, decreasing=T)[c(1:3,7)]], lty=3)
```
