The $k$-th moment exists only when the shape parameter $\alpha$ is greater than $k$. So the higher the alpha value, the less fat the tail (?)

```
n <- 1000
sam <- 100

m <- matrix(0,n,sam)

locat = 10
alpha = 10

for(i in 1:n) m[i,] <- rpareto(sam,location = locat, shape = alpha)

mat1 = m
mat1[1,] <- m[which(m == max(m), arr.ind = TRUE)[1],]
mat1[which(m == max(m), arr.ind = TRUE)[1],] <- m[1,]
```

```
set.seed(2021)
par(mfrow=c(1,2))
options(repr.plot.width=15, repr.plot.height=8)

plot(density(mat1[1,]), xlim=c(7,20), ylim=c(0,.8),
     xlab='', ylab='', main=paste('Pareto sampling with shape (alpha) =', alpha),
     col=rgb(runif(1,.5,1),0,runif(1,.6,1),0.005))

for (i in 1:n) lines(density(mat1[i,]),
          col=rgb(runif(1,.5,1),0,runif(1,.6,1),0.005))

data <- sort(as.numeric(mat1[1,]))
ypoints <- 1 - ppoints(data)
plot(data, ypoints, log='xy', cex=.4, main=paste('Pareto sampling with shape (alpha) =', alpha),
     xlab='x in log scale', ylab='1 - F(x) in log scale',
     col=rgb(runif(1,.5,1),0,runif(1,.8,1),0.005), ylim=c(0.005,2))

for (i in 2:n){
  data <- sort(as.numeric(mat1[i,]))
  ypoints <- 1 - ppoints(data)
  lines(data,ypoints,
  col=rgb(runif(1,.5,1),0,runif(1,.6,1),0.005))
}
```

```
n <- 1000
sam <- 100

m <- matrix(0,n,sam)

ave = 10
sd = 1

for(i in 1:n) m[i,] <- rnorm(sam, mean=ave , sd)

mat4 = m
mat4[1,] <- m[which(m == max(m), arr.ind = TRUE)[1],]
mat4[which(m == max(m), arr.ind = TRUE)[1],] <- m[1,]




par(mfrow=c(1,2))
options(repr.plot.width=15, repr.plot.height=8)

plot(density(mat4[1,]), ylim=c(0, 1.5*max(density(mat4[1,])$y)),
     xlab='', ylab='', main=paste('Normal sampling with mean =', ave, ' and SD = ', sd),
     col=rgb(runif(1,.5,1),0,runif(1,.6,1),0.005))

for (i in 1:n) lines(density(mat4[i,]),
          col=rgb(runif(1,.5,1),0,runif(1,.6,1),0.005))

data <- sort(as.numeric(mat4[1,]))
ypoints <- 1 - ppoints(data)
plot(data, ypoints, log='xy', cex=.4, main=paste('Normal sampling with mean =', ave, ' and SD = ', sd),
     xlab='x in log scale', ylab='1 - F(x) in log scale',
     col=rgb(runif(1,.5,1),0,runif(1,.8,1),0.005), ylim=c(0.005,2))

for (i in 2:n){
  data <- sort(as.numeric(mat4[i,]))
  ypoints <- 1 - ppoints(data)
 lines(data, ypoints,col=rgb(runif(1,.5,1),0, runif(1,0.9,1), 0.02), xlim=c(0,log(max(mat4))))
}
```
