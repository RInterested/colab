```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))

u = 1
v = -1

#creating the quiver plot
plt.figure(figsize=(10, 10))
plt.quiver(x,y,u,v, color=(0.9, 0.2, 0.9, 0.9) , width=0.01, headwidth=1, headlength=1.5, scale=25)

#Displaying the plot
plt.show()
```

```python
x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))

u = x
v = y

color = x**2+y**2
plt.rcParams["image.cmap"] = "gnuplot"
plt.figure(figsize=(10, 10))
plt.quiver(x,y,u,v, color, width=.004)
plt.show()
```

```python
x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

u = -y/np.sqrt(x**2 + y**2)
v = x/np.sqrt(x**2 + y**2)
plt.figure(figsize=(10, 10))
plt.quiver(x,y,u,v)
plt.show()
```

```python
x,y = np.meshgrid(np.arange(-2, 2, .2), np.arange(-2, 2, .25))
z = x*np.exp(-x**2 - y**2)
v, u = np.gradient(z)
n=-2
color = np.sqrt(v**2 + u**2)
plt.rcParams["image.cmap"] = "seismic"
plt.figure(figsize=(10, 10))
plt.quiver(x,y,u,v,color,width=.008)
plt.show()
```

```python
lim = 10
int = 0.22 *lim
x,y = np.meshgrid(np.arange(-lim, lim, int), np.arange(-lim, lim, int))

u = 3 * np.cos(np.arctan2(y,x)) - np.sqrt(x**2+y**2) * np.sin(np.arctan2(y,x))
v = 3 * np.sin(np.arctan2(y,x)) + np.sqrt(x**2+y**2) * np.cos(np.arctan2(y,x))

color = x**2 + y**2
plt.rcParams["image.cmap"] = "Greys_r"
mult = 1
plt.figure(figsize=(mult*lim, mult*lim))
plt.quiver(x,y,u,v,color, linewidths=.006, lw=.1)
plt.show()
```

```python
lim = 9
int = 0.18 *lim
x,y = np.meshgrid(np.arange(-lim, lim, int), np.arange(-lim, lim, int))

# contract in the center
a = 0.2  # how far to contract
b = 0.7  # how strongly to contract
c = 1 - b*np.exp(-((0.4*x/lim)**2 + (0.4*y/lim)**2)/a**2)
x, y = c*x, c*y

u = 10 * np.cos(np.arctan2(y,x)) -  np.sqrt(x**2+y**2) * np.sin(np.arctan2(y,x))
v = 10 * np.sin(np.arctan2(y,x)) +  np.sqrt(x**2+y**2) * np.cos(np.arctan2(y,x))

color = np.exp(0.2*np.sqrt(x**2 + y**2))
plt.rcParams["image.cmap"] = "hsv_r"
mult = 1
plt.figure(figsize=(mult*lim, mult*lim))
plt.quiver(x,y,u,v,color, width=0.0085, headwidth=3, headlength=4, scale=145)
plt.show()
```

```python
lim = 10
N = 10

theta = np.linspace(0, 2*np.pi, N*2)

stretch_factor = 1.5
r = np.linspace(1, lim, N)**stretch_factor

R, THETA = np.meshgrid(r, theta)
x = R * np.cos(THETA)
y = R * np.sin(THETA)

# x,y = np.meshgrid(x, y)
r = x**2 + y**2

u = 3 * np.cos(np.arctan2(y,x)) - np.sqrt(r) * np.sin(np.arctan2(y,x))
v = 3 * np.sin(np.arctan2(y,x)) + np.sqrt(r) * np.cos(np.arctan2(y,x))

plt.rcParams["image.cmap"] = "Oranges_r"
mult = 1
plt.figure(figsize=(mult*lim, mult*lim))
plt.quiver(x,y,u,v,r, width=.003)
```

```python
from mpl_toolkits.mplot3d import axes3d

%matplotlib inline

fig = plt.figure(figsize=(15,15))

ax = fig.gca(projection='3d')

x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) *
     np.sin(np.pi * z))

ax.quiver(x, y, z, u, v, w, length=0.1)

plt.show()
```

```python
# 1D arrays
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

# Meshgrid
X,Y = np.meshgrid(x,y)

# Assign vector directions
Ex = (X + 1)/((X+1)**2 + Y**2) - (X - 1)/((X-1)**2 + Y**2)
Ey = Y/((X+1)**2 + Y**2) - Y/((X-1)**2 + Y**2)

# Depict illustration
plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,Ex,Ey, density=1.4, linewidth=None, color='#A23BEC')
plt.plot(-1,0,'-or')
plt.plot(1,0,'-og')
plt.title('Electromagnetic Field')

# Show plot with grid
plt.grid()
plt.show()
```
