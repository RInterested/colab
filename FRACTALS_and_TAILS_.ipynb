```
data(danish)
```

```
options(repr.plot.width=9, repr.plot.height=7)
emplot(danish, 'xy', pch=19, col='sienna', cex=.4)
```

```
d <- sort(danish)
```

```
n1 <- vector()
for(i in 1:(length(d)/2)) n1[i] <- d[i]+d[length(d)-i+1]
```

```
n2 <- vector()
for(i in 1:(length(d)-1)) n2[i] <- d[i]+d[i+1]
```

```
emplot(danish,'xy', type='l', col='navyblue', lwd=3)

data <- sort(n1)
ypoints <- 1 - ppoints(data)
lines(data, ypoints,
     xlab='x in log scale', ylab='1 - F(x) in log scale', col='red4', lwd=3)

data <- sort(n2)
ypoints <- 1 - ppoints(data)
lines(data, ypoints,
     xlab='x in log scale', ylab='1 - F(x) in log scale', col='orange2', lwd=3)
```

```
set.seed(2021)
locat1 = 10
alpha1 = 10
m <- rpareto(1e4, location = locat1, shape = alpha1)


d <- sort(as.numeric(m))
ypoints <- 1 - ppoints(d)
plot(d, ypoints, log='xy', cex=.4, main=paste('Pareto sampling with shape (alpha) =', alpha1, 'and loc = ', locat1),
     xlab='x in log scale', ylab='1 - F(x) in log scale', type='l', col='navyblue', lwd=3)

m1 <- vector()
for(i in 1:(length(m)/2)) m1[i] <- d[i]+d[length(m)-i+1]
data <- sort(m1)
ypoints <- 1 - ppoints(data)
lines(data, ypoints, col='red4', lwd=3)

m2 <- vector()
for(i in 1:(length(m)-1)) m2[i] <- d[i]+d[i+1]
data <- sort(m2)
ypoints <- 1 - ppoints(data)
lines(data, ypoints, col='orange2', lwd=3)
```

```
set.seed(2021)
locat2 = 10
alpha2 = 1
m <- rpareto(1e4,location = locat2, shape = alpha2)


d <- sort(as.numeric(m))
ypoints <- 1 - ppoints(d)
plot(d, ypoints, log='xy', cex=.4, main=paste('Pareto sampling with shape (alpha) =', alpha2, 'and loc =', locat2),
     xlab='x in log scale', ylab='1 - F(x) in log scale', type='l', col='navyblue', lwd=3)

m3 <- vector()
for(i in 1:(length(m)/2)) m3[i] <- d[i]+d[length(m)-i+1]
data <- sort(m3)
ypoints <- 1 - ppoints(data)
lines(data, ypoints, col='red', lwd=3)


m4 <- vector()
for(i in 1:(length(m)-1)) m4[i] <- d[i]+d[i+1]
data <- sort(m4)
ypoints <- 1 - ppoints(data)
lines(data, ypoints, col='chartreuse3', lwd=4)
```

```
par(mfrow=c(1,2))
options(repr.plot.width=10, repr.plot.height=5)
h1 <- d[1:(.5*length(d))]
h2 <- d[(.5*length(d)+1):(length(d))]

h2prime <- h2[1:which.min(abs(h2 - (h2[1] *(ceiling(h1[length(h1)] / h1[1])))))]

ypoints1 <- dpareto(h1,location = locat2, shape = alpha2)
plot(h1, ypoints1, type='l', lwd=3, xlab='x', ylab='f(x)', col='orange3')

ypoints2prime <- dpareto(h2prime,location = locat2, shape = alpha2)
plot(h2prime, ypoints2prime, type='l', lwd=3, xlab='x', ylab='f(x)', col='orange3')

ypoints1 <- dpareto(h1,location = locat2, shape = alpha2)
plot(h1, ypoints1, type='l', lwd=3, log='xy')

ypoints2 <- dpareto(h2,location = locat2, shape = alpha2)
plot(h2, ypoints2, type='l', lwd=3, log='xy')
```

```
locat3 = 10
scale3 = 2

m <- rcauchy(1e4, location = locat3, scale = scale3)


d <- sort(as.numeric(m))
ypoints <- 1 - ppoints(d)


options(repr.plot.width=9, repr.plot.height=8)
plot(d, ypoints, log='xy', cex=.4, main=paste('Cauchy sampling with scale =', scale3, 'and loc =', locat3),
     xlab='x in log scale', ylab='1 - F(x) in log scale', type='l', col='navyblue', lwd=3)

cau1 <- vector()
for(i in 1:(length(m)/2)) cau1[i] <- d[i]+d[length(m)-i+1]
data <- sort(cau1)
ypoints <- 1 - ppoints(data)
lines(data, ypoints, col='red4', lwd=3)

cau2 <- vector()
for(i in 1:(length(m)-1)) cau2[i] <- d[i]+d[i+1]
data <- sort(cau2)
ypoints <- 1 - ppoints(data)
lines(data, ypoints, col='orange2', lwd=3)
```

```
par(mfrow=c(1,2))
options(repr.plot.width=10, repr.plot.height=5)

d <- d[d>10]
k1 <- d[1:(.5*length(d))]
k2 <- d[(.5*length(d)+1):(length(d))]

k2prime <- k2[1:which.min(abs(k2 - (k2[1] *(ceiling(k1[length(k1)] / k1[1])))))]

y1 <- dcauchy(k1, location = locat3, scale = scale3)
plot(k1, y1, type='l', lwd=3,  xlab='x', ylab='f(x)', col='red3', main='Cauchy')

y2prime <- dcauchy(k2prime, location = locat3, scale = scale3)
plot(k2prime, y2prime, type='l', lwd=3,  xlab='x', ylab='f(x)', col='red3', main='Cauchy')

y1 <- dcauchy(k1, location = locat3, scale = scale3)
plot(k1, y1, type='l', lwd=3, log='xy')

y2prime <- dcauchy(k2prime, location = locat3, scale = scale3)
plot(k2prime, y2prime, type='l', lwd=3, log='xy')
```
