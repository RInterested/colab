```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
pic = cv2.imread('pic.jpeg',0)
plt.imshow(pic, cmap = 'gray')
```

```python
# fft
f = np.fft.fft2(pic)
# shift the center
f_shifted = np.fft.fftshift(f)

freq = np.log(np.abs(f_shifted))
plt.imshow(abs(freq), cmap='gray')
```

```python
# inverse fft
f_ishift = np.fft.ifftshift(f_shifted)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)
plt.imshow(img_back, cmap = 'gray')
```

```python
v = f_shifted
n = 2
matv = v[:,np.arange(1,np.shape(v)[1],n)] # Takes out every other row off the frequencies.
print(matv.shape)
freqv = np.log(np.abs(matv))
plt.imshow(freqv, cmap='gray')
```

```python
# inverse fft
fz_ishift = np.fft.ifftshift(matv)
img_back_z = np.fft.ifft2(matv)
img_back_z = np.abs(img_back_z)
plt.imshow(img_back_z, cmap = 'gray')
```
