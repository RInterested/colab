```
install.packages("sfsmisc")
library(sfsmisc)
```

```
options(repr.plot.width=8, repr.plot.height=6)
# Gamble: Toss a fair coin: if H wealth increases by 50 pct; T wealth decreases by 40 pct.

# Wealth of one person over 10 tosses:

set.seed(2021)
tosses <- 10
princ <- 1
wealth <- c(princ, princ * cumprod(ifelse(rbinom(tosses,1,0.5)==1,1.5,0.6)))
plot(wealth, type='l', ylim=c(0.1,10), log='y', yaxt='n',
      main='Wealth of one person 10 tosses', xlab='tosses')
eaxis(2)
#axis(2,c(0.1,1,10))
abline(h=1)
```

```
options(repr.plot.width=10, repr.plot.height=8)
princ <- 100
people <- 1e6 # Now for 1 million people (black line)
ensemble <- matrix(0,people, tosses+1)
for(i in 1:people) ensemble[i,] <- c(princ, princ * cumprod(ifelse(rbinom(tosses,1,0.5)==1,1.5,0.6)))
matplot(t(ensemble[1:23,]), type='l',
        ylim=c(min(ensemble[1:23,]),max(ensemble[1:23,])),
        log='y', yaxt='n',
        main= 'Ensemble wealth over 23 people (and summary of 1 mil.) ',
        xlab = 'tosses', ylab='wealth')
abline(h=100)
lines(colMeans(ensemble[1:23,]), col='firebrick', lwd=5, lty=3)
lines(colMeans(ensemble), lwd=5)
eaxis(2)
legend(0,10^-1, c('Pointwise average of 23 people', 'Ensemble of 1 million people'),
              lty = c(3,1),
              lwd=5, col = c('firebrick',1),  bty='n')
```

```
options(repr.plot.width=8, repr.plot.height=6)
lives <- 10
tosses <- 1e4
matr <- matrix(0, lives, tosses+1)
for(i in 1:lives) matr[i,] <- c(princ, princ * cumprod(ifelse(rbinom(tosses,1,0.5)==1,1.5,0.6)))
plot(matr[1,], type='l', log='y', yaxt='n', ylab='wealth', ylim=c(min(matr), max(matr)),
      main=paste('Wealth of one person',tosses, 'tosses and', lives, 'lives.'), xlab='tosses')
abline(h=1)
eaxis(2)
for(i in 1:nrow(matr)) lines(matr[i,], col=i)
```
