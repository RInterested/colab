### Geodesic Equation:

From [here](https://youtu.be/faoiWD8yedk).


Following calculus of variations, the shortest curve from a point $\vec x_1$ to $\vec x_2$ is assumed to exist and be $\vec y(\lambda),$ such that $\vec y(\lambda_1)=\vec x_1$ and $\vec y(\lambda_2)=\vec x_2.$

Consider and arbitrary curve $\vec \eta(\lambda),$ such as $\eta(\lambda_1)=\eta(\lambda_2)=0.$

The curve $\vec y(\lambda) + \varepsilon \eta(\lambda) $ is  infinitesimally close to $\vec y(\lambda).$

The arc length of a curve in general parameterized by $\lambda$ is

$$s =\int\sqrt{g_{ij}dx^i dx^j}= \int_{\lambda_1}^{\lambda_2}\sqrt{g_{ij}\,\vec x(\lambda) \dot x^i \dot x^j}d\lambda$$

For the curve $\vec y(\lambda) + \varepsilon \eta(\lambda) $

$$s(\vec y + \varepsilon \eta)=\int_{\lambda_1}^{\lambda_2}\sqrt{g_{ij}\,(\vec y + \varepsilon \vec\eta)\, (\dot y^i + \varepsilon \dot\eta^i) \, (\dot y^j + \varepsilon \dot\eta^j)}d\lambda$$

The Taylor expansion up to the first order of $\varepsilon$ results in

$$s(\vec y + \varepsilon \eta)= s(\vec y) + \int_{\lambda_1}^{\lambda_2} \varepsilon \left[\frac{\partial g_{ij}(\vec y)}{\partial y^k}\, \eta^k \dot y^i \dot y^j + \color{blue}{g_{ij}(\vec y) \dot \eta^i \dot y^j} + \color{red}{ g_{ij}(\vec y) \dot y^i \dot \eta^j  }\right]\frac{1}{\sqrt{g_{ij}\vec y(\lambda)\dot y^i \dot y^j}} d\lambda$$

Parameterizing by arclength $\lambda = s$ the expression $\frac{1}{\sqrt{g_{ij}\vec y(\lambda)\dot y^i \dot y^j}}$ becomes one because

$ds =\sqrt{g_{ij}dx^i dx^j} ds.$

Doing integration by parts of $\color{blue}{g_{ij}(\vec y) \dot \eta^i \dot y^j}$:

$$\int_{s_1}^{s_2}\color{blue}{g_{ij}(\vec y) \dot \eta^i \dot y^j}ds=\int_{s_1}^{s_2}\left [\frac{d}{ds}(g_{ij}\eta^i\dot y^j) - \dot g_{ij}\eta^i\dot y^j- g_{ij}\eta^i \ddot y^j  \right ]ds$$

where $\int uv = uv - \int vdu$ corresponds to $u= g_{ij}\dot y$ and $v=\eta.$

The first term is $0$ because of $\eta$ at $s_1$ and $s_2$ is $0.$ This leaves

$$\int_{s_1}^{s_2}\color{blue}{g_{ij}(\vec y) \dot \eta^i \dot y^j}ds=\int_{s_1}^{s_2}\left [- \dot g_{ij}\dot y^j- g_{ij} \ddot y^j  \right ]\eta^i \, ds$$

Similarly,

$$\int_{s_1}^{s_2}\color{red}{ g_{ij}(\vec y) \dot y^i \dot \eta^j }=\int_{s_1}^{s_2}\left [- \dot g_{ij}\dot y^i - g_{ij} \ddot y^i  \right ]\eta^i \, ds$$


Therefore,

$$s(\vec y + \varepsilon \eta)= s(\vec y) + \int_{s_1}^{s_2} \varepsilon \left[\frac{\partial g_{ij}(\vec y)}{\partial y^k}\, \eta^k \dot y^i \dot y^j - \dot g_{ij}\dot y^j- g_{ij} \ddot y^j - \dot g_{ij}\dot y^i - g_{ij} \ddot y^i  \eta^i \right] ds$$

with a change of dummy variable:

$$s(\vec y + \varepsilon \eta)= s(\vec y) + \int_{s_1}^{s_2} \varepsilon \left[\frac{\partial g_{ij}(\vec y)}{\partial y^k}\, \dot y^i \dot y^j - \dot g_{ij}\dot y^j- g_{ij} \ddot y^j - \dot g_{ij}\dot y^i - g_{ij} \ddot y^i  \right]\, \eta^k ds$$

The integral must be $0,$ and given that $\eta$ is arbritary, it must be that

$$\frac{\partial g_{ij}(\vec y)}{\partial y^k}\, \dot y^i \dot y^j - \dot g_{ij}\dot y^j- g_{ij} \ddot y^j - \dot g_{ij}\dot y^i - g_{ij} \ddot y^i  = 0$$

Recognizing the following identities:

$$\begin{align}
\dot g_{kj}\dot y^j &=\frac{\partial g_{kj}(\vec y)}{\partial y^i}\dot y^i \dot y^j\\[2ex]\dot g_{ik}\dot y^i &=\frac{\partial g_{ik}(\vec y)}{\partial y^i}\dot y^i \dot y^j\\[2ex]
g_{kj}\ddot y^j &=  g_{ik}\ddot y^i
\end{align}$$

results in following identity:

$$g_{kj}\ddot y^j = -\frac 1 2 \left[ \frac{\partial g_{kj}(\vec y)}{\partial y^i} + \frac{\partial g_{ik}(\vec y)}{\partial y^i} - \frac{\partial g_{ij}(\vec y)}{\partial y^k}    \right]\, \dot y^i \dot y^j$$

Now

$$(g_{qk})^{-1} g_{kj}\ddot y^j = -\frac 1 2 (g_{qk})^{-1} \left[ \frac{\partial g_{kj}(\vec y)}{\partial y^i} + \frac{\partial g_{ik}(\vec y)}{\partial y^i} - \frac{\partial g_{ij}(\vec y)}{\partial y^k}    \right]\, \dot y^i \dot y^j$$

results in

$$\ddot y^q = -\frac 1 2 (g_{qk})^{-1} \left[ \frac{\partial g_{kj}(\vec y)}{\partial y^i} + \frac{\partial g_{ik}(\vec y)}{\partial y^i} - \frac{\partial g_{ij}(\vec y)}{\partial y^k}    \right]\, \dot y^i \dot y^j$$

which identifies the Christoffel symbol:

$$\ddot y^q = -\Gamma^q_{ij}\, \dot y^i \dot y^j$$

resulting in a [non-linear](https://youtu.be/bVy1F5T8QE4) second order equation for every component $q$.
